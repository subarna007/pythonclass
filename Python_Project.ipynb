```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('https://www.worldometers.info/coronavirus/', 'html.parser')
soup = requests.get(url)
soup
```

```python
soup = BeautifulSoup(soup.text, "lxml")
table_code = soup.table
table_code
```

```python
soup = BeautifulSoup(soup.text, "lxml")
table_code = soup.table
table_code
```

```python
tags = table_code.find_all('tr')
tags
```

```python
data = []
for i in tags:
  y = i.text.split('\n')[1:]
  if y[0] != "":
    data.append(y)

```

```python
import csv
file = open('covid.csv','w')
x = csv.writer(file)
x.writerows(data)
file.close()
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
url = 'https://www.worldometers.info/coronavirus/'
req = requests.get(url)
page = BeautifulSoup(req.content, 'html.parser')

table = page.find_all('table',id="main_table_countries_today")[0]

df = pd.read_html(str(table), displayed_only=False)[0]
df.head(5)
```
