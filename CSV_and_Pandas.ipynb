```python
import csv
csv_columns = ['name','age','address']
data = [
    {'name': 'Subarna', 'age': '23', 'address': 'Lalitpur'},
    {'name': 'Rewanta', 'age': '22', 'address': 'Canada'},
    {'name': 'Puskar', 'age': '24', 'address': 'Australia'}
   ]

csvfile = open('data.csv','w')
writer = csv.DictWriter(csvfile, fieldnames = csv_columns)
writer.writeheader()
writer.writerows(data)
csvfile.close()
```
