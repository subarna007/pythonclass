```python
a = int(input("Enter value of a = "))
b = int(input("Enter value of b = "))
o = input("Enter + or - or * or / ")
def add(a,b):
    return a+b

def sub(a,b):
    return a-b

def mul(a,b):
    return a*b

def div(a,b):
    return a/b

def cal(a,b,o):
    if o == '+':
        print(add(a,b))

    elif o == '-':
        print(sub(a,b))

    elif o == '*':
        print(mul(a,b))

    elif (o == '/') and (b != 0):
        print(div(a,b))

    elif (o == '/') and (b != 0): #and operator used case scenario
        print("The value of b cannot be zero when o = / ")

    else:
        print(" Error ! Enter valid operator")

cal(a,b,o)
```

```python
bill = str()
all_total = 0
n = int(input("Enter n = "))

def billing_system(n):
    global bill,all_total
    for i in range (n):
        name = input("Enter name = ")
        price = int(input("Enter price = "))
        quantity = float(input("Enter quantity = "))
        total = price * quantity
        bill = bill + f"{name} {price} {quantity} {total} \n"
        all_total = all_total + total
    return bill,all_total


bill,all_total = billing_system(n)
print(bill)
print(all_total)
```

```python
#bill format using recursive function
bill = []
def billing_system():
    global bill
    name = input("Enter name = ")
    price = int(input("Enter price = "))
    quantity = float(input("Enter quantity = "))
    total = price * quantity
    bill = bill + [[name,price,quantity,total]]
    x = input("Press Enter for more entry ")
    if x == '':
        billing_system()

billing_system()
```
